###머신러닝이란

-> 컴퓨터에 학습능력을 부여하는 것을 목적, 데이터가 가지는 규칙성 분명히 함

###머신러닝 종류

1. **지도학습**
 - 정답 데이터를 얻을 수 있는 문제를 다루는 학습, 예측 결과가 맞는지 평가 (정규선형모델, 일반선형모델 등)

2. **비지도학습**
 - 정답 데이터를 얻을 수 없는 문제를 다루는 학습

3. **강화학습**
 - 주어진 상황에서 이득이 최대가 되는 행동을 찾는 문제

4. **룰베이스**
 - 사람이 미리 룰 지정하고 그 룰에 따라서 예측 결과 출력

###통계학과 머신러닝 구분

-> 통계모델의 목적은 데이터를 얻는 프로세스 이해

-> 머신러닝의 목적은 미지의 데이터를 계산해서 획득하는 것

###정규화, 리지 회귀, 라소 회귀

1. **정규화**
 - 파라미터를 추정할 때 손실함수에 벌칙항을 도입합으로써 계수가 큰 값이 되는 것을 막는 기법
 - 벌칙항을 정규화항이라고도 함
 - 파라미터의 축소추정이라고도 함

2. **리지 회귀(L2 정규화)**
 - 정규화항으로 계수의 제곱합을 이용한 회귀모델
 - 잔차제곱합을 작게 하고 싶지만 벌칙은 받고 싶지 않다. 그래서 절대치가 작은 계수가 추정된다.

3. **라소회귀(L1 정규화)**
 - 정규화항으로 계수의 절댓값의 합을 이용한 회귀모델
 - 벌칙항이 잔차제곱합이 아닌 절댓값의 합이라는거 외엔 리지회귀와 동일

=> **a 결정하는 방법**: a값을 여러가지로 변화시켜 교차검증법으로 가장 잘 예측한 값 채택

=> **독립변수 표준화**: 정규화 전 독립변수를 평균 0, 표준편차 1로 표준화 해야

###과학습 방지 방법

1. AIC 이용해 미지의 파라미터 수 줄여 모델 단순화

2. 정규화로 벌칙항 삽입해 계수 절대치 줄여 독립이 종속에 미치는 영향 감소
(라소에서는 계수의 대부분이 0이 되는 변수 선택하는 셈)

###정규화의 의의

1. **통계모델**: AIC 등을 이용한 변수 선택의 결점 완화 가능 (AIC는 교차검증으로 계산량 많지만 리지, 라소는 적은 계산 비용으로 가능)

2. **머신러닝**: 과학습 방지, AIC 계산과 변수 선택 동시에 할 수 있음
